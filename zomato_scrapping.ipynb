```python
def append_json_to_csv(json_data, csv_file_path):
    # Check if the CSV file exists
    file_exists = os.path.isfile(csv_file_path)

    # Open the CSV file in append mode
    with open(csv_file_path, mode='a', newline='') as file:
        writer = csv.writer(file)

        # If file does not exist, write the header
        if not file_exists:
            header = json_data.keys()  # assuming json_data is a dictionary
            writer.writerow(header)

        # Write the JSON data to CSV
        writer.writerow(json_data.values())
```

```python
def get_res_details(response_text: str, res_id: int , res_url: str = None):
  # Regex pattern to capture the JSON inside the window.__PRELOADED_STATE__
  pattern = r'window\.__PRELOADED_STATE__\s*=\s*JSON\.parse\("(?P<json_data>.*?)"\);'

  # Search for the pattern in the response text
  match = re.search(pattern, response_text)

  if match:
      # Extract the matched JSON data (it's escaped, so unescape it)
      json_data = match.group('json_data').encode('utf-8').decode('unicode_escape')

      # Parse the JSON string into a Python dictionary
      preloaded_state = json.loads(json_data)
      # res = preloaded_state.get('pages', {}).get('restaurant',{})
      res_contact = preloaded_state.get('pages', {}).get('restaurant',{}).get(f'{res_id}',{}).get('sections', {}).get('SECTION_RES_CONTACT', {})
      if res_contact:
        res_contact.update(res_contact.get('phoneDetails'))
        res_contact.pop('phoneDetails', None)
        res_contact.update({'res_id': res_id, 'url': res_url})
        append_json_to_csv(res_contact, '/content/drive/MyDrive/zomato/res_contact.csv')
        # with open(f'/content/drive/MyDrive/zomato/preloaded_state_{res_id}.json', 'w') as file:
        #   json.dump(res_contact, file, indent=4)

        return res_contact
  else:
      return False
```

```python
def restaurant_details(res_id):
  response = requests.get(f'http://zoma.to/r/{res_id}', cookies=generate_cookies(), headers=headers)
  if response.ok:
    if 'restaurants' not in response.url:
      print('url:', response.url)
      return get_res_details(response.text, res_id, response.url)
    else:
      print('url_fail:', response.url)
      return False
```
